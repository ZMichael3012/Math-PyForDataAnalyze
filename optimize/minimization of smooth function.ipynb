```
import numpy as np
import scipy.optimize as opt
import matplotlib.pyplot as plt
%matplotlib inline
```

```
f = lambda x: np.sin( x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
```

```
x = np.arange(1, 31)
```

```
res = opt.minimize(f, 25)
```

```
res
```

```
plt.plot(x, f(x))
```

```
res = opt.minimize(method = 'BFGS', x0 = 2, fun = f)
```

```
res
```
