```
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as opt
%matplotlib inline
```

```
f = lambda x: np.sin(x / 5.0) * np.exp(x / 10.0) + 5 * np.exp(-x / 2.0)
```

```
x = np.arange(1, 31)
```

```
plt.plot(x, f(x))
```

```
res = opt.differential_evolution(func = f, bounds = [(1, 30)])
```
