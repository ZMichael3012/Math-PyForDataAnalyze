```python
import numpy as np
import re
from scipy import spatial
```

```python
sentences = []
```

```python
with open('sentences.txt') as f:
    sentences = f.readlines()
```

```python
for sen in sentences:
    print(sen)
```

```python
low_sens = [re.split('[^a-z]', sen.lower()) for sen in sentences]
```

```python
for sen in low_sens:
    while('' in sen):
        sen.remove('')
```

```python
for sen in low_sens:
    print(sen)
```

```python
dif_words = {}
```

```python
i = 0
for sen in low_sens:
    for word in sen:
        if word not in dif_words:
            dif_words[word] = i
            i += 1
```

```python
len(sentences), len(dif_words)
```

```python
matrix = np.zeros((len(sentences), len(dif_words)))
```

```python
j = 0
for word in dif_words:
    for i in range(matrix.shape[0]):
        amount = float(low_sens[i].count(word))
        matrix[i, j] = amount
    j += 1
```

```python
values = []
```

Calculating cosine distance between the first sentence and the 
others

```python
for i in range(1, matrix.shape[0]):
    values.append(spatial.distance.cosine(matrix[0], matrix[i]))
```
