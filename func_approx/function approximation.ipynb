```
%matplotlib inline
from scipy import linalg
import numpy as np
import matplotlib.pyplot as plt
```

```
x = np.arange(1, 16)
print(x)
```

```
f = lambda x: np.sin(x / 5) * np.exp(x / 10) + 5 * np.exp(-x / 2)
```

```
plt.plot(x, f(x))
plt.show()
```

```
x0 = [1, 4, 10, 15]
```

```
matrix = np.array([[i**j for j in range(4)] for i in x0])
```

```
vec = np.array([f(i) for i in x0])
```

```
print(matrix)
print(vec)
```

```
polyn = linalg.solve(matrix, vec)
```

```
print(polyn)
```
